```python
 """
********************************************
* IPL 20082022 BALL-BY-BALL ANALYSIS      *
* Author : Vimaladitya                     *
* Profile: Data Analyst | Python Developer *
********************************************
---------------------------------------------------------
|This project analyzes 15 seasons of IPL data including: |
|- Top Batsmen, Bowlers, Fielders                        |
|- Boundary & Strike Analysis                            |
|- Powerplay, Middle, Death Overs Performance            |
|- Team-wise Scoring Patterns                            | 
|- Dismissal Breakdown & Extras Analysis                 |
---------------------------------------------------------
"""
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("IPL_Ball_By_Ball_2008_2022.csv")
# stats of batsmen(2008 to 2022)
print("*********** RECORDS OF BATSMAN ************")
bat_six = df[df["batsman_run"]==6]
bat_four = df[df["batsman_run"]==4]
bat_sin = df[df["batsman_run"]==1]
bat_dou = df[df["batsman_run"]==2]
bat_th = df[df["batsman_run"]==3]
bat_fiv = df[df["batsman_run"]==5]
bat_score = df["batsman_run"].sum()
#pie plot of runs scored by batsmen in ipl
x = ("singles","doubles","triples","four","six")
y = (len(bat_sin),len(bat_dou),len(bat_th),len(bat_four),len(bat_six))
plt.pie(y,labels = x,autopct='%1.1f%%')
plt.title("RUNS SCORED BY BATSMAN")
plt.show()
print("Total number of singles taken by Batsman",len(bat_sin))
print("Total number of doubles taken by Batsman",len(bat_dou))
print("Total number of tribles taken by Batsman",len(bat_th))
print("Total number of four's taken by Batsman",len(bat_four))
print("Total number of six's taken by Batsman",len(bat_six))
print("Total number of five taken by Batsman",len(bat_fiv))
print("total runs scored by the batsman",bat_score)
# stats of top 5 BATSMEN from (2008 to 2022)
print("*********** TOP 5 BATSMEN  RUN SCORERS***********")
top_batsmen = df.groupby("batter")["batsman_run"].sum().sort_values(ascending=False).head(5)
print(top_batsmen)
# stats of bowlers from (2008 to 2022)
print("*********** RECORDS OF BOWLERS ************")
bat_wicket = df[df["isWicketDelivery"]==1]
bat_caught = df[df["kind"]=="caught"]
bat_runout = df[df["kind"]=="run out"]
bat_bowled = df[df["kind"]=="bowled"]
bat_stumped = df[df["kind"]=="stumped"]
bat_cab = df[df["kind"]=="caught and bowled"]
bat_lbw = df[df["kind"]=="lbw"]
print("Total number of wickets taken by Bowlers",len(bat_wicket))
print("Total number of caught outs",len(bat_caught))
print("Total number of run outs",len(bat_runout))
print("Total number of bowled outs",len(bat_bowled))
print("Total number of stumpings",len(bat_stumped))
print("Total number of caught and bowled",len(bat_cab))
print("Total number of lbw",len(bat_lbw))
#pie chart for batsman dismssal methods from 2008 t0 2022
x = ("caught out","run out","stumped out","bowled out","caught and bowled","LBW")
y = (len(bat_caught),len(bat_runout),len(bat_bowled),len(bat_stumped),len(bat_cab),len(bat_lbw))
plt.figure(figsize=(7,4))
plt.pie(y,labels=x,autopct='%1.1f%%')
plt.title(" BATSMAN DISMISSALS METHODS ")
plt.show()
# runs given as extras(2008 to 2022)
print("************* RUNS GIVEN AS EXTRAS **************")
extras_df = df[df["extra_type"].notna()]
ex_leg_runs    = extras_df[extras_df["extra_type"] == "legbyes"]["total_run"].sum()
ex_wide_runs   = extras_df[extras_df["extra_type"] == "wides"]["total_run"].sum()
ex_noball_runs = extras_df[extras_df["extra_type"] == "noballs"]["total_run"].sum()
total_extra_runs = extras_df["total_run"].sum()
print("Total extra runs:", total_extra_runs)
print("Leg Byes:", ex_leg_runs)
print("Wides:", ex_wide_runs)
print("No Balls:", ex_noball_runs)
#bar chat for runs given as extras
x = ["Leg Byes", "Wides", "No Balls"]
y = [ex_leg_runs, ex_wide_runs, ex_noball_runs]
plt.figure(figsize=(7,4))
plt.bar(x, y, color=["green", "red", "blue"])
plt.title("Runs Conceded as Extras")
plt.xlabel("Type of Extra")
plt.ylabel("Runs")
plt.show()
# stats of top 5 bowlers(2008 to 2022)
print("*********** TOP 5 BOWLERS ***********")
top_bowlers = df[(df["isWicketDelivery"] == 1) & (df["kind"] != "run out")]
top_bowlers = top_bowlers.groupby("bowler").size().sort_values(ascending=False).head(5)
print(top_bowlers)
# stats of top 5 BOUNDARY HITTERS(2008 to 2022)
print("*********** TOP 5 BOUNDARY HITTERS ***********")
fours = df[df["batsman_run"] == 4].groupby("batter").size()
sixes = df[df["batsman_run"] == 6].groupby("batter").size()
boundary_hitters = (fours.add(sixes, fill_value=0)).sort_values(ascending=False).head(5)
print(boundary_hitters)
# stats of top 5 FIELDERS(2008 to 2022)
print("*********** TOP 5 FIELDERS ***********")
top_fielders = df[df["kind"] == "caught"].groupby("fielders_involved").size().sort_values(ascending=False).head(5)
print(top_fielders)
# stats of the 5 POWER PLAY TOP SCORES(2008 to 2022)
print("************* POWER PLAY TOP SCORES *****************")
powerplay = df[df["overs"]<=6].groupby("batter")["batsman_run"].sum().sort_values(ascending = False).head(5)
print(powerplay)
# stats of  the top 5 MIDDLE OVER scorers(2008 to 2022)
print("************** MIDDLE OVER TOP SCORERS ****************")
middleovers = df[(df["overs"] >= 7) & (df["overs"] <= 15)].groupby("batter")["batsman_run"].sum().sort_values(ascending = False).head(5)
print(middleovers)
# stats of  the top 5 DEATH OVER  scorers(2008 to 2022)
print("***************DEATH OVER TOP SCORERS*****************")
deathover = df[df['overs']>15].groupby("batter")['batsman_run'].sum().sort_values(ascending= False).head(5)
print(deathover)
# stats of  the top 5 DEATH OVER  scorers(2008 to 2022)
print("************* TOP 5 TEAM WHICH HAS HIGH POWER PLAY SCORES***************")
powerplay_team = df[df["overs"]<=6].groupby("BattingTeam")["batsman_run"].sum().sort_values(ascending = False).head(5)
print(powerplay_team)
# stats of  the top 5 MIDDLE OVER scorers(2008 to 2022)
print("************ TOP 5 TEAM WHICH HAS HIGH MIDDLE OVER SCORES *****************")
middleovers_team = df[(df["overs"] >= 7) & (df["overs"] <= 15)].groupby("BattingTeam")["batsman_run"].sum().sort_values(ascending = False).head(5)
print(middleovers_team)
# stats of  the top 5 HIGH DEATH OVER SCORES(2008 to 2022)
print("***************TOP 5 TEAM WHICH HAS HIGH DEATH OVER SCORES*****************")
deathover_team = df[df['overs']>15].groupby("BattingTeam")['batsman_run'].sum().sort_values(ascending= False).head(5)
print(deathover_team)
# stats of  the top 5 POWERPLAY BOWLERS(2008 to 2022)
print("*************** TOP 5 POWERPLAY BOWLERS ****************")
powerplay_bowler = df[(df["overs"] <= 6) & (df["isWicketDelivery"] == 1) & (df["kind"] != "run out")]
powerplay_bowler = powerplay_bowler.groupby("bowler").size().sort_values(ascending=False).head(5)
print(powerplay_bowler)
# stats of  the top 5 MIDDLE OVER BOWLERS(2008 to 2022)
print("************** TOP 5 MIDDLE OVER BOWLERS ****************")
midover_bowler = df[(df["overs"] >= 7) & (df["overs"] <= 15) & (df["isWicketDelivery"] == 1) & (df["kind"] != "run out")]
midover_bowler = midover_bowler.groupby("bowler").size().sort_values(ascending=False).head(5)
print(midover_bowler)
# stats of  the top 5 DEATH OVER BOWLERS(2008 to 2022)
print("***************TOP 5 DEATH OVER BOWLERS ***************")
deathover_bowler = df[(df["overs"] > 15) & (df["isWicketDelivery"] == 1) & (df["kind"] != "run out")]
deathover_bowler = deathover_bowler.groupby("bowler").size().sort_values(ascending=False).head(5)
print(deathover_bowler)
"""
---------------------------------------------------------
                     CONCLUSION
---------------------------------------------------------

This project gives a clear overview of IPL performance
from 20082022. It highlights the top batsmen, bowlers,
fielders, high-scoring teams, and scoring patterns across
Powerplay, Middle Overs, and Death Overs.

Basic visualizations and statistics help us understand how
runs are made, how wickets fall, and which players and
teams perform consistently across different match phases.

This analysis provides a simple but complete summary of
IPL trends using Python, Pandas, and Matplotlib.
---------------------------------------------------------
"""

```
